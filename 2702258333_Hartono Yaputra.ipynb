Split Data into Train and Test Data

```python
idf=df.drop(['loan_status'],axis=1)
odf=df['loan_status']
```

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(idf, odf, test_size = 0.2, random_state = 42)
```

Handle Missing Values

Diasumsikan person_income yang bernilai NULL merupakan nilai 0 alias orang tersebut belum memiliki income sesuai dengan data saat ini

```python
x_train['person_income'].fillna(0, inplace=True)
x_test['person_income'].fillna(0, inplace=True)
```

```python
x_train['person_gender'].replace('fe male',"female", inplace=True)
x_train['person_gender'].replace('Male',"male", inplace=True)

x_test['person_gender'].replace('fe male',"female", inplace=True)
x_test['person_gender'].replace('Male',"male", inplace=True)
```

```python
encode_gender={"person_gender": {"male":1,"female" :0}}
x_train=x_train.replace(encode_gender)
x_test=x_test.replace(encode_gender)
```

```python
encode_pl={"previous_loan_defaults_on_file": {"Yes":1,"No" :0}}
x_train=x_train.replace(encode_pl)
x_test=x_test.replace(encode_pl)
```

```python
from sklearn.preprocessing import OneHotEncoder

pe_train_enc=x_train[['person_education']]
ph_train_enc=x_train[['person_home_ownership']]
li_train_enc=x_train[['loan_intent']]


pe_test_enc=x_test[['person_education']]
ph_test_enc=x_test[['person_home_ownership']]
li_test_enc=x_test[['loan_intent']]

pe_enc=OneHotEncoder()
ph_enc=OneHotEncoder()
li_enc=OneHotEncoder()

pe_train_enc=pd.DataFrame(pe_enc.fit_transform(pe_train_enc).toarray(),columns=pe_enc.get_feature_names_out())
ph_train_enc=pd.DataFrame(ph_enc.fit_transform(ph_train_enc).toarray(),columns=ph_enc.get_feature_names_out())
li_train_enc=pd.DataFrame(li_enc.fit_transform(li_train_enc).toarray(),columns=li_enc.get_feature_names_out())

pe_test_enc=pd.DataFrame(pe_enc.fit_transform(pe_test_enc).toarray(),columns=pe_enc.get_feature_names_out())
ph_test_enc=pd.DataFrame(ph_enc.fit_transform(ph_test_enc).toarray(),columns=ph_enc.get_feature_names_out())
li_test_enc=pd.DataFrame(li_enc.fit_transform(li_test_enc).toarray(),columns=li_enc.get_feature_names_out())

x_train=x_train.reset_index()
x_test=x_test.reset_index()

x_train_enc=pd.concat([x_train,pe_train_enc,ph_train_enc,li_train_enc], axis=1)
x_test_enc=pd.concat([x_test,pe_test_enc,ph_test_enc,li_test_enc], axis=1)
```

Scaling for person_income and loan_amnt

```python
from sklearn.preprocessing import StandardScaler

features_to_scale = ['person_income', 'loan_amnt']

scaler = StandardScaler()

x_train_scaled = x_train_enc.copy()
x_train_scaled[features_to_scale] = scaler.fit_transform(x_train_enc[features_to_scale])

x_test_scaled = x_test_enc.copy()
x_test_scaled[features_to_scale] = scaler.transform(x_test_enc[features_to_scale])
```

```python
x_train_scaled.head()
```

Modeling using XGBoost Classifier and Random Forest Classifier

```python
from sklearn.ensemble import RandomForestClassifier
import warnings
warnings.filterwarnings('ignore')

rfc = RandomForestClassifier()
rfc.fit(x_train_scaled, y_train)
```

```python
rfcpred = rfc.predict(x_test_scaled)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, rfcpred, target_names=['Test','Pred']))
```

```python
from xgboost import XGBClassifier

xgbc = XGBClassifier()
xgbc.fit(x_train_scaled,y_train)
```

```python
xgbpred = xgbc.predict(x_test_scaled)
```

```python
print(classification_report(y_test, xgbpred,target_names=['Test','Pred']))
```
